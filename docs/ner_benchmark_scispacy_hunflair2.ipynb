```python
import torch
import flair
import spacy
from flair.nn import Classifier
from flair.data import Sentence
from datasets import load_dataset
from seqeval.metrics import classification_report
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
flair.device = device
```

```python
def reconstruct_text_and_spans(tokens):
    text = ""
    spans = []
    for tok in tokens:
        start = len(text)
        text += tok
        end = len(text)
        spans.append((start, end))
        text += " "
    return text.strip(), spans


def bio_tags_from_ids(tag_ids, id2label):
    return [id2label[i] for i in tag_ids]


def align_predictions_to_tokens(predicted_entities, token_spans, label_map, overlap_threshold=0.5):
    tags = ["O"] * len(token_spans)

    for ent in predicted_entities:
        mapped_label = label_map.get(ent["label"])
        if mapped_label is None:
            continue

        first_token_in_span = True
        for i, (tok_start, tok_end) in enumerate(token_spans):
            overlap = max(0, min(ent["end"], tok_end) - max(ent["start"], tok_start))
            tok_len = tok_end - tok_start
            if tok_len > 0 and (overlap / tok_len) >= overlap_threshold:
                tags[i] = f"B-{mapped_label}" if first_token_in_span else f"I-{mapped_label}"
                first_token_in_span = False

    return tags
```

```python
def run_scispacy(nlp, text):
    doc = nlp(text)
    return [
        {"text": ent.text, "label": ent.label_, "start": ent.start_char, "end": ent.end_char}
        for ent in doc.ents
    ]


def run_hunflair2(hunflair, text):
    sentence = Sentence(text)
    hunflair.predict(sentence)
    return [
        {
            "text": ent.text,
            "label": ent.get_label("ner").value,
            "start": ent.start_position,
            "end": ent.end_position,
        }
        for ent in sentence.get_spans("ner")
    ]


def spans_from_bio(tokens, bio_tags):
    spans = []
    current_tokens, current_label = [], None
    for tok, tag in zip(tokens, bio_tags):
        if tag.startswith("B-"):
            if current_tokens:
                spans.append((" ".join(current_tokens), current_label))
            current_tokens, current_label = [tok], tag[2:]
        elif tag.startswith("I-") and current_label == tag[2:]:
            current_tokens.append(tok)
        else:
            if current_tokens:
                spans.append((" ".join(current_tokens), current_label))
            current_tokens, current_label = [], None
    if current_tokens:
        spans.append((" ".join(current_tokens), current_label))
    return spans
```

```python
def print_sample_comparisons(gold_dataset, gold_id2label, model_runners, num_samples=10, seed=0):
    import random
    random.seed(seed)
    indices = random.sample(range(len(gold_dataset)), min(num_samples, len(gold_dataset)))

    for idx in indices:
        example = gold_dataset[idx]
        tokens = example["tokens"]
        gold_tags = bio_tags_from_ids(example["tags"], gold_id2label)
        gold_spans = spans_from_bio(tokens, gold_tags)
        text, token_spans = reconstruct_text_and_spans(tokens)

        print(f"\n{'-' * 90}")
        print(f"TEXT: {text}")
        print(f"{'-' * 90}")
        print(f"  [GOLD]      {gold_spans if gold_spans else '(no entities)'}")

        for model_name, (run_fn, label_map) in model_runners.items():
            predicted_entities = run_fn(text)
            pred_tags = align_predictions_to_tokens(predicted_entities, token_spans, label_map)
            pred_spans = spans_from_bio(tokens, pred_tags)
            print(f"  [{model_name}]  {pred_spans if pred_spans else '(no entities)'}")
```

```python
def benchmark_model(model_name, run_fn, gold_dataset, gold_id2label, label_map, max_examples=200):
    y_true, y_pred = [], []

    n = min(max_examples, len(gold_dataset))
    for i in range(n):
        example = gold_dataset[i]
        tokens = example["tokens"]
        gold_tags = bio_tags_from_ids(example["tags"], gold_id2label)

        text, token_spans = reconstruct_text_and_spans(tokens)
        predicted_entities = run_fn(text)
        pred_tags = align_predictions_to_tokens(predicted_entities, token_spans, label_map)

        y_true.append(gold_tags)
        y_pred.append(pred_tags)

    print(f"\n{'=' * 80}\n{model_name}  (evaluated on {n} examples)\n{'=' * 80}")
    print(classification_report(y_true, y_pred, digits=3))
```

```python
from huggingface_hub import hf_hub_download
import pandas as pd
from datasets import Dataset as HFDataset

print("\nLoading gold dataset (bigbio/bionlp_st_2013_cg) via direct parquet read...", flush=True)
parquet_path = hf_hub_download(
    repo_id="bigbio/bionlp_st_2013_cg",
    repo_type="dataset",
    revision="refs/convert/parquet",
    filename="bionlp_st_2013_cg_source/train/0000.parquet",
)
cg_df = pd.read_parquet(parquet_path)
cg_gold = HFDataset.from_pandas(cg_df)
print(f"Loaded {len(cg_gold)} CG documents.", flush=True)
```

```python
CG_LABEL_MAP_BIONLP13CG = {
    "GENE_OR_GENE_PRODUCT": "Gene_or_gene_product",
    "TISSUE": "Tissue",
    "ORGAN": "Organ",
    "CANCER": "Cancer",
    "CELLULAR_COMPONENT": "Cellular_component",
    "ORGANISM": "Organism",
    "CELL": "Cell",
    "ANATOMICAL_SYSTEM": "Anatomical_system",
    "ORGANISM_SUBDIVISION": "Organism_subdivision",
    "DEVELOPING_ANATOMICAL_STRUCTURE": "Developing_anatomical_structure",
    "ORGANISM_SUBSTANCE": "Organism_substance",
    "PATHOLOGICAL_FORMATION": "Pathological_formation",
    "SIMPLE_CHEMICAL": "Simple_chemical",
    "AMINO_ACID": "Amino_acid",
    "IMMATERIAL_ANATOMICAL_ENTITY": "Immaterial_anatomical_entity",
    "MULTI-TISSUE_STRUCTURE": "Multi-tissue_structure",
}
```

```python
def whitespace_tokenize_with_spans(text):
    tokens, spans = [], []
    i = 0
    for tok in text.split():
        i = text.index(tok, i)
        tokens.append(tok)
        spans.append((i, i + len(tok)))
        i += len(tok)
    return tokens, spans
```

```python
def bigbio_doc_to_gold_bio(doc, label_map):
    text = doc["text"]
    tokens, token_spans = whitespace_tokenize_with_spans(text)

    gold_entities = []
    for ann in doc["text_bound_annotations"]:
        start, end = ann["offsets"][0]
        ann_text = ann["text"][0]
        gold_entities.append({
            "text": ann_text,
            "label": label_map.get(ann["type"], ann["type"]),
            "start": int(start),
            "end": int(end),
        })

    identity_map = {v: v for v in label_map.values()}
    gold_tags = align_predictions_to_tokens(gold_entities, token_spans, identity_map)
    return text, tokens, token_spans, gold_tags
```

```python
def benchmark_bionlp13cg_on_cg(nlp_bio, cg_dataset, label_map, max_examples=100):
    y_true, y_pred = [], []
    n = min(max_examples, len(cg_dataset))
    for i in range(n):
        doc = cg_dataset[i]
        text, tokens, token_spans, gold_tags = bigbio_doc_to_gold_bio(doc, label_map)
        predicted_entities = run_scispacy(nlp_bio, text)
        pred_tags = align_predictions_to_tokens(predicted_entities, token_spans, label_map)
        y_true.append(gold_tags)
        y_pred.append(pred_tags)

    print(f"\n{'=' * 80}\nscispaCy bionlp13cg vs gold BioNLP-13-CG  (evaluated on {n} docs)\n{'=' * 80}")
    print(classification_report(y_true, y_pred, digits=3))
```

```python
print("\n" + "#" * 90)
print("# SAMPLE COMPARISON — BioNLP-13-CG (bionlp13cg's real training corpus)")
print("#" * 90)
sample_doc = cg_gold[0]
sample_text, sample_tokens, sample_spans, sample_gold_tags = bigbio_doc_to_gold_bio(
    sample_doc, CG_LABEL_MAP_BIONLP13CG
)
print(f"\nTEXT (truncated): {sample_text[:300]}...")
print(f"  [GOLD]         {spans_from_bio(sample_tokens, sample_gold_tags)}")
sample_pred = run_scispacy(nlp_bio, sample_text)
sample_pred_tags = align_predictions_to_tokens(sample_pred, sample_spans, CG_LABEL_MAP_BIONLP13CG)
print(f"  [bionlp13cg]   {spans_from_bio(sample_tokens, sample_pred_tags)}")
```
